```python
import pandas as pd
import numpy as np
import pandas_ta as pta
import plotly.express as px
import matplotlib.pyplot as plt
import plotly.graph_objs as go
from plotly.subplots import make_subplots
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/DatumLearning/Single_Stock_Analysis/main/TCS.csv")
```

```python
data.head()
```

```python
data.drop(["Symbol" , "Series" , "Prev Close" , "Last" , "VWAP" , "Turnover" ,
           "Trades" , "Deliverable Volume" , "%Deliverble"] , axis = 1 , inplace = True)
```

```python
#Simple Moving Average
```

```python
data["sma50"] = data.ta.sma(length = 50)
data["sma200"] = data.ta.sma(length = 200)
```

```python
help(data.ta.sma)
```

```python
dir(data.ta)
```

```python
data.loc[150:250]
```

```python
data['sma50']
```

```python
buy_signal = (data['sma50'] > data['sma200']) & (data['sma50'] < data['sma200']).shift(1)
sell_signal = (data['sma50'] < data['sma200']) & (data['sma50'] > data['sma200']).shift(1)
data["buy_signal"] = buy_signal
data["sell_signal"] = sell_signal
```

```python
data = pd.concat([data , data.ta.bbands(length = 20)] , axis = 1)
```

```python
help(data.ta.bbands)
```

```python
data
```

```python
#Initial graph
fig = make_subplots(rows = 1 , cols = 1 , shared_xaxes = True)

#Add stock prices and sma50 lines
stock_price_line = go.Scatter(x = data.Date , y = data.Close , mode = 'lines' , name = 'Stock price')
sma50_line = go.Scatter(x = data.Date , y = data.sma50 , mode = 'lines' , name = 'SMA50')
sma200_line = go.Scatter(x = data.Date , y = data.sma200 , mode = 'lines' , name = 'SMA200')
buy_markers = go.Scatter(x = data[data["buy_signal"] == 1].Date , y = data[data["buy_signal"] == 1]['sma50'] ,
                         mode = 'markers' , marker=dict(color='black', size=5))
sell_markers = go.Scatter(x = data[data["sell_signal"] == 1].Date , y = data[data["sell_signal"] == 1]['sma50'] ,
                          mode = 'markers' , marker=dict(color='red', size=5))
fig.add_trace(stock_price_line)
fig.add_trace(sma50_line)
fig.add_trace(sma200_line)
fig.add_trace(buy_markers)
fig.add_trace(sell_markers)
fig.show()
```
