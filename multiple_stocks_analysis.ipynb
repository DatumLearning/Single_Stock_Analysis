```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
```

```python
hdfc = pd.read_csv("https://raw.githubusercontent.com/DatumLearning/Single_Stock_Analysis/main/hdfc.csv")
tcs = pd.read_csv("https://raw.githubusercontent.com/DatumLearning/Single_Stock_Analysis/main/TCS.csv")
reli = pd.read_csv("https://raw.githubusercontent.com/DatumLearning/Single_Stock_Analysis/main/RELIANCE.csv")
ongc = pd.read_csv("https://raw.githubusercontent.com/DatumLearning/Single_Stock_Analysis/main/ONGC.csv")
```

```python
data = pd.concat([hdfc['Close'] , tcs['Close'] , ongc['Close'] , reli['Close']] ,
                axis = 1)
data.columns = ["HDFC" , "tcs" , "ONGC" , "RELI"]
scatter_matrix(data , figsize = (3.5 , 3.5) , hist_kwds = {'bins' : 250})
plt.show()
```

```python
#Volatility
hdfc['returns'] = (hdfc['Close']/hdfc['Close'].shift(1)) -1
tcs['returns'] = (tcs['Close']/tcs['Close'].shift(1)) -1
reli['returns'] = (reli['Close']/reli['Close'].shift(1)) -1
ongc['returns'] = (ongc['Close']/ongc['Close'].shift(1)) -1
```

```python
hdfc.shape[0]
```

```python
names = ['hfcs' , 'tcs' , 'reli' , 'ongc']
plt.bar(x = names , height = [hdfc.shape[0] , tcs.shape[0] , reli.shape[0] , ongc.shape[0]])
plt.show()
```

```python
out = np.hstack([hdfc['returns'] , tcs['returns'] , reli['returns'] , ongc['returns']])
max_scale = np.nanmax(out)
max_scale
```

```python
plt.hist(hdfc['returns'] , bins = 100 , alpha = 0.5 , label = "hdfc")
plt.hist(tcs['returns'] , bins = 100 , alpha = 0.5 , label = "tcs")
plt.hist(reli['returns'] , bins = 100 , alpha = 0.5 , label = "reli")
plt.hist(ongc['returns'] , bins = 100 , alpha = 0.5 , label = "ongc")
plt.xlim(-max_scale , max_scale)
plt.legend()
plt.show()
```
